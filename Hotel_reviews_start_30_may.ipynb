```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

def tokenize_and_lowercase(text):
    tokens = word_tokenize(text)
    # Convert tokens to lowercase
    tokens_lower = [token.lower() for token in tokens]
    return tokens_lower

# Assuming Hotel_df is your DataFrame
Hotel_df.loc[:, 'Negative_Review_tokenized'] = Hotel_df['Negative_Review'].apply(tokenize_and_lowercase)
Hotel_df.loc[:, 'Positive_Review_tokenized'] = Hotel_df['Positive_Review'].apply(tokenize_and_lowercase)
```

```python
from nltk.corpus import stopwords

def remove_stop_words(tokens):
    stop_words = set(stopwords.words('english'))
    negation_words = ['not', 'no', "isn't", "doesn't", "can't", "won't", "shouldn't", "wouldn't", "didn't", "doesnt", "cant", "wont", "shouldnt", "wouldnt", "didnt"]

    # Remove negation words from stop_words
    stop_words.difference_update(negation_words)

    filtered_tokens = []
    i = 0
    while i < len(tokens):
        if tokens[i].lower() in negation_words and i + 1 < len(tokens):
            # Check if the next word is in the list of stop words
            if tokens[i+1].lower() not in stop_words:
                filtered_tokens.append(tokens[i] + ' ' + tokens[i+1])
                i += 1
            else:
                filtered_tokens.append(tokens[i] + tokens[i+1])
                i += 1
        else:
            filtered_tokens.append(tokens[i])
        i += 1

    return filtered_tokens
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]
```

```python
def lemmatize_tokens(tagged_tokens):
    lemmatized_tokens = []
    for token, pos_tag in tagged_tokens:
        pos = pos_tag[0].lower() if pos_tag else 'n'
        pos = pos if pos in ['a', 'r', 'n', 'v'] else 'n'
        lemmatized_tokens.append(lemmatizer.lemmatize(token, pos))
    return lemmatized_tokens
```

```python
Hotel_df.loc[:, 'Negative_Review_tokenized_SWR_POS_Lem'] = Hotel_df['Negative_Review_tokenized_SWR_POS'].apply(lemmatize_tokens)
Hotel_df.loc[:, 'Positive_Review_tokenized_SWR_POS_Lem'] = Hotel_df['Positive_Review_tokenized_SWR_POS'].apply(lemmatize_tokens)
```
